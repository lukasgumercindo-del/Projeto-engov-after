```python
import pandas as pd
import numpy as np

np.random.seed(42)

datas = pd.date_range(start="2020-01-01", end="2026-06-30", freq="D")

produtos_info = {
    "Citrus": {"ano_lancamento": 2020, "preco_base": 7.50},
    "Tangerina": {"ano_lancamento": 2020, "preco_base": 7.50},
    "Berry Vital": {"ano_lancamento": 2022, "preco_base": 8.00},
    "Red Energy": {"ano_lancamento": 2022, "preco_base": 8.00},
    "Mental Drink Brabo": {"ano_lancamento": 2024, "preco_base": 8.90}
}

registros = []

for data in datas:
    ano = data.year
    mes = data.month
    
    fator_sazonal = 1.0
    if mes == 2:
        fator_sazonal = 2.5
    elif mes == 12:
        fator_sazonal = 1.8
    elif mes in [1, 3, 7]:
        fator_sazonal = 1.3

    for produto, info in produtos_info.items():
        if ano >= info["ano_lancamento"]:
            crescimento_ano = 1 + (ano - 2020) * 0.15
            volume_base = np.random.randint(500, 1500)
            
            quantidade_vendida = int(volume_base * fator_sazonal * crescimento_ano)
            
            preco_unitario = round(info["preco_base"] * np.random.uniform(0.95, 1.08), 2)
            faturamento_total = round(quantidade_vendida * preco_unitario, 2)
            
            registros.append({
                "data": data,
                "produto": f"Engov After {produto}",
                "sabor": produto,
                "quantidade_vendida": quantidade_vendida,
                "preco_unitario": preco_unitario,
                "faturamento_total": faturamento_total
            })

df_vendas = pd.DataFrame(registros)
df_vendas.to_csv("vendas_engov_after_2020_2026.csv", index=False)

print(f"Base de dados gerada com sucesso! Total de linhas: {df_vendas.shape[0]}")
```

```python
import pandas as pd 

df = pd.read_csv("vendas_engov_after_2020_2026.csv")
df.head().style.set_table_styles(
    [{'selector': 'th', 'props': [('background-color', "#ddd3d3"), ('color', '#333'), ('font-weight', 'bold')]},]
).format({'preco_unitario': 'R${:.2f}', 'faturamento_total': 'R${:.2f}'})
```

```python
import plotly.express as px 

faturamento_por_sabor = df.groupby('sabor')['faturamento_total'].sum().sort_values(ascending=False).reset_index()

fig_faturamento_por_sabor = px.bar(
    faturamento_por_sabor,
    x = 'sabor',
    y = 'faturamento_total', 
    title = 'Faturamento Total por Sabor (Engov After 2020-2026)',
    labels = {'sabor': 'Sabor', 'faturamento_total': 'Faturamento Total (R$)'},
    color = 'faturamento_total',
    color_continuous_scale = 'Blues'
)

fig_faturamento_por_sabor.update_layout(
    yaxis_tickprefix='R$',
    yaxis_tickformat=',.2f',
    title_x=0.5
)
fig_faturamento_por_sabor.show()
```

```python
valor_por_sabor = df.groupby('sabor')['quantidade_vendida'].sum().sort_values(ascending=False).reset_index()

fig_volume = px.bar(
    valor_por_sabor,
    x='sabor',
    y='quantidade_vendida',
    title='Volume Total de Unidades Vendidas por Sabor (2020-2026)',
    labels={'sabor': 'Sabor do Produto', 'quantidade_vendida': 'Total de Unidades Vendidas'},
    color='quantidade_vendida',
    color_continuous_scale='Greens'
)
fig_volume.show()

fig_faturamento_por_sabor.update_layout(
    yaxis_tickprefix='R$',
    yaxis_tickformat=',.2f',
    title_x=0.5
)
fig_faturamento_por_sabor.show()
```

```python
df['data'] = pd.to_datetime(df['data'])
df['ano'] = df['data'].dt.year

faturamento_por_ano = df.groupby('ano')['faturamento_total'].sum().reset_index()

fig_temporal = px.line(
    faturamento_por_ano, 
    x = 'ano',
    y = 'faturamento_total',
    title = 'Evolução do Faturamento Anual de Vendas (2020-2026)',
    labels = {'ano': 'Ano Comcercial', 'faturamento_total': 'Faturamento Total (R$)'},
    markers = True
) 
fig_temporal.update_traces(line_color=' #FF5733', line_width=3)
fig_temporal.update_layout( yaxis_tickprefix='R$',
    yaxis_tickformat=',.2f',
    title_x=0.5)
fig_temporal.show()
```
